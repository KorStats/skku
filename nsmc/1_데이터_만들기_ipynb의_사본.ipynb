```python
import pandas as pd
df=pd.read_csv('/content/gdrive/MyDrive/졸업논문/DATA/NSMC_train.csv', encoding='utf-8-sig')
```

```python
df.head(3)
```

```python
import pandas as pd
samples = {1 : 5000, 0 :4000 }
sampled_data = pd.DataFrame()
for category, n in samples.items():
    category_data = df[df['label'] == category]
    sampled_category_data = category_data.sample(n=n, random_state=42)
    sampled_data = pd.concat([sampled_data, sampled_category_data], ignore_index=True)
sampled_data.reset_index(drop=True, inplace=True)
sampled_data.head(3)
```

```python
sampled_data['label'].value_counts()
```

```python
sampled_data.to_csv('NSMC_train_불균형_0.8.csv', encoding='utf-8-sig')
```

```python
df=sampled_data
```
