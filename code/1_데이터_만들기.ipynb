```python
import pandas as pd
df=pd.read_csv('/content/gdrive/MyDrive/신문기사_train_original.csv', encoding='utf-8-sig')
```

```python
df['길이'] = df['text'].apply(len)
```

```python
df=df[df['길이']<1000]
df['category'].value_counts()
```

```python
import pandas as pd
samples = {'종합': 4000,'경제': 2000,'사회': 2000, '정치' : 2000, '스포츠' : 1000, 'IT,과학' : 500, '교육/입시/NIE' : 500}
sampled_data = pd.DataFrame()
for category, n in samples.items():
    category_data = df[df['category'] == category]
    sampled_category_data = category_data.sample(n=n, random_state=42)
    sampled_data = pd.concat([sampled_data, sampled_category_data], ignore_index=True)
sampled_data.reset_index(drop=True, inplace=True)
sampled_data.head(3)
```

```python
sampled_data['category'].value_counts()
```

```python
sampled_data.to_csv('신문기사_단문.csv', encoding='utf-8-sig')
```
