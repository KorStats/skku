```python
import pandas as pd
df=pd.read_csv('/content/gdrive/MyDrive/★졸업논문/신문기사_train_original.csv', encoding='utf-8-sig')
```

```python
df['category'].value_counts()
```

```python
import pandas as pd
samples = {'정치': 5000,'경제': 4000,'사회': 2500,'IT,과학' : 500, '교육/입시/NIE' : 100}
sampled_data = pd.DataFrame()
for category, n in samples.items():
    category_data = df[df['category'] == category]
    sampled_category_data = category_data.sample(n=n, random_state=42)
    sampled_data = pd.concat([sampled_data, sampled_category_data], ignore_index=True)
sampled_data.reset_index(drop=True, inplace=True)
sampled_data.head(3)
```

```python
sampled_data['category'].value_counts()
```

```python
sampled_data.to_csv('신문기사_불균형데이터_train.csv', encoding='utf-8-sig')
```

```python
remaining_data = df[~df['id'].isin(sampled_data['id'])]
```

```python
remaining_data['category'].value_counts()
```

```python
import pandas as pd
samples = {'정치': 1000,'경제': 1000,'사회': 1000,'IT,과학' : 1000, '교육/입시/NIE' : 1000}
sampled_data = pd.DataFrame()
for category, n in samples.items():
    category_data = remaining_data[remaining_data['category'] == category]
    sampled_category_data = category_data.sample(n=n, random_state=42)
    sampled_data = pd.concat([sampled_data, sampled_category_data], ignore_index=True)
sampled_data.reset_index(drop=True, inplace=True)
sampled_data.head(3)
```

```python
sampled_data['category'].value_counts()
```

```python
sampled_data.to_csv('신문기사_불균형데이터_test.csv', encoding='utf-8-sig')
```
