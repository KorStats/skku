```python
data_list = []
for q, label in zip(df['document'], df['label'])  :
    data = []
    data.append(q)
    data.append(str(label))
    data_list.append(data)

#from sklearn.model_selection import train_test_split
#dataset_train, dataset_test = train_test_split(data_list, test_size=0.25, random_state=0, shuffle=True)

train=pd.DataFrame(data_list)

data_list = []
for q, label in zip(df_test['document'], df_test['label'])  :
    data = []
    data.append(q)
    data.append(str(label))
    data_list.append(data)

test=pd.DataFrame(data_list)

train.rename(columns = {0 : 'text', 1 : 'category'}, inplace = True)
test.rename(columns = {0 : 'text', 1 : 'category'}, inplace = True)
```

```python
def clean_korean_documents(documents):
    # í…ìŠ¤íŠ¸ ì •ì œ (íŠ¹ìˆ˜ê¸°í˜¸ë§Œ ì œê±°)
    for i, document in tqdm(enumerate(documents), total=len(documents), desc="Cleaning Special Characters"):
        document = re.sub(r'[^\w\sã„±-ã…£ê°€-í£]', '', document)  # íŠ¹ìˆ˜ê¸°í˜¸ë§Œ ì œê±°, ì •ê·œ í‘œí˜„ì‹
        documents[i] = document
    return documents

def remove_words_with_numbers(text):
    return ' '.join([word for word in text.split(',') if not re.search(r'\d', word)])

# í•œê¸€ì´ ì•„ë‹Œ ë¬¸ìë¥¼ ì œê±°í•˜ëŠ” í•¨ìˆ˜
def remove_non_korean(text):
    return re.sub(r'[^ã„±-ã…£ê°€-í£\s]', '', text)
```

```python
train['text'] = clean_korean_documents(train['text'])
train['text'] = train['text'].apply(lambda x: ','.join(x.split()))
train['text'] = train['text'].apply(remove_words_with_numbers)
train['text'] = train['text'].apply(remove_non_korean)

test['text'] = clean_korean_documents(test['text'])
test['text'] = test['text'].apply(lambda x: ','.join(x.split()))
test['text'] = test['text'].apply(remove_words_with_numbers)
test['text'] = test['text'].apply(remove_non_korean)
```

```python
# 3. í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ì—ì„œ í•„ìš”í•œ ì—´ ì¶”ì¶œ ë° ë¼ë²¨ ì¸ì½”ë”©
X_train = train['text']
y_train = train['category']
X_test = test['text']
y_test=test['category']

# 4. ë¼ë²¨ ì¸ì½”ë”©
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded=label_encoder.fit_transform(y_test)
num_labels = len(label_encoder.classes_)
```

```python
# 5. KoBERT ëª¨ë¸ì„ ìœ„í•œ í† í¬ë‚˜ì´ì € ë° ë°ì´í„°ì…‹ ìƒì„±
model_name = 'monologg/kobert'
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
```

```python
# í…ìŠ¤íŠ¸ë¥¼ í† í°í™”í•˜ëŠ” í•¨ìˆ˜ ì •ì˜
def tokenize_function(examples):
    return tokenizer(examples['text'], padding='max_length', truncation=True, max_length=256)
```

```python
# ë°ì´í„°ì…‹ ìƒì„± ë° í† í¬ë‚˜ì´ì¦ˆ
train_df = pd.DataFrame({'text': X_train, 'label': y_train_encoded})
test_df = pd.DataFrame({'text': X_test, 'label': y_test_encoded})
train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(test_df)
train_dataset = train_dataset.map(tokenize_function, batched=True)
test_dataset = test_dataset.map(tokenize_function, batched=True)
```

```python
# PyTorch í…ì„œ í˜•ì‹ìœ¼ë¡œ ë³€í™˜
train_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'label'])
test_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'label'])

# 5. KoBERT ëª¨ë¸ ì„¤ì • ë° í•™ìŠµ íŒŒë¼ë¯¸í„° ìµœì í™”
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels).to(device)
```

```python
#CPU ì¸¡ì •

import psutil
import time
from transformers import Trainer, TrainingArguments

# CPU ì‚¬ìš©ëŸ‰ ë° ì†ë„ ì¸¡ì • í•¨ìˆ˜
def measure_cpu_usage_and_time(train_function):
    start_time = time.time()  # ì‹œì‘ ì‹œê°„ ê¸°ë¡
    start_cpu = psutil.cpu_percent(interval=None)  # ì‹œì‘ CPU ì‚¬ìš©ëŸ‰

    train_function()  # í•™ìŠµ í•¨ìˆ˜ í˜¸ì¶œ

    end_time = time.time()  # ì¢…ë£Œ ì‹œê°„ ê¸°ë¡
    end_cpu = psutil.cpu_percent(interval=None)  # ì¢…ë£Œ CPU ì‚¬ìš©ëŸ‰

    elapsed_time = end_time - start_time
    avg_cpu_usage = (start_cpu + end_cpu) / 2

    print(f"Elapsed Time: {elapsed_time:.2f} seconds")
    print(f"Average CPU Usage: {avg_cpu_usage:.2f}%")
```

```python
import optuna
from transformers import Trainer, TrainingArguments
from sklearn.metrics import f1_score
import numpy as np

# `compute_metrics` í•¨ìˆ˜
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    f1 = f1_score(labels, predictions, average='macro')  # F1 ì ìˆ˜ ê³„ì‚°
    return {'eval_f1': f1}  # 'eval_f1' ë°˜í™˜

# Optunaë¥¼ í™œìš©í•œ í•˜ì´í¼íŒŒë¼ë¯¸í„° íŠœë‹
def model_init():
    return AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)
```

```python
training_args = TrainingArguments(
    output_dir='/content',
    num_train_epochs=10,  # ì—í¬í¬ ìˆ˜ ê³ ì •
    per_device_train_batch_size=64,
    learning_rate=5.0e-05,  # ìµœì í™”ëœ í•™ìŠµë¥ 
    weight_decay=0.01,  # ìµœì í™”ëœ weight decay
    evaluation_strategy='epoch',
    save_strategy='epoch',
    logging_dir='/content',
    logging_steps=10,
    fp16=True,
    optim="adamw_torch",  # AdamW fused ì ìš© (PyTorch 2.0 ì´ìƒ)
    disable_tqdm=False,
    load_best_model_at_end=True,
    metric_for_best_model="f1",  # ê°€ì¥ ì¢‹ì€ ëª¨ë¸ì„ accuracy ê¸°ì¤€ìœ¼ë¡œ ì„ íƒ
    greater_is_better=True,  # accuracyê°€ ë†’ì€ ëª¨ë¸ì„ ì €ì¥
    save_total_limit=2,  # ì €ì¥í•  ëª¨ë¸ ê°œìˆ˜ ì œí•œ
    report_to="none"
)
```

```python
# ìµœì ì˜ íŒŒë¼ë¯¸í„°ë¡œ Trainer êµ¬ì„±
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)

trainer7 = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,  # ì „ì²´ í•™ìŠµ ë°ì´í„°ë¡œ í‰ê°€
    compute_metrics=compute_metrics
)
```

```python
# CPU ì‚¬ìš©ëŸ‰ ë° ì†ë„ ì¸¡ì •
measure_cpu_usage_and_time(trainer7.train)
```
